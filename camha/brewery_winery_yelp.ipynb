```python
import requests
import pandas as pd
import json
from pprint import pprint
import api_keys
import time 
```

```python
brewery = pd.read_csv('brewery_yelp.csv', encoding='utf-8')
brewery.rename({'brewery names': 'names'}, axis=1, inplace=True)
brewery.head()
```

```python
winery = pd.read_csv('winery_yelp.csv', encoding='utf-8')
winery.rename({'winery names': 'names'}, axis=1, inplace=True)
winery.head()
```

```python
brewery_winery = pd.concat([brewery,winery],sort=True, axis=0)
brewery_winery.head()
```

```python
brewery_winery.to_csv('brewery_winery_yelp.csv', encoding='utf-8')

brewery_winery.head()
```
